```python
#importing all dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score , StratifiedKFold
from sklearn.svm import SVC
```

```python
#getting the dataset
dataset = pd.read_csv(r"C:\Users\Akhilesh Kumar\Downloads\diabetes_prediction_dataset.csv")
dataset.head(3)
```

```python
#checking shape of dataset
dataset.shape
```

```python
#checking null values in the dataset
dataset.isnull().sum()
```

```python
#checking different types of data types in smoking_history column of dataset
dataset['smoking_history'].value_counts()
```

```python
#checking different types of data types in gender column of dataset
dataset['gender'].value_counts()
```

```python
#Preprocessing the data columns smoking_history and gender to convert it into numerical columns
le = LabelEncoder()
# 0 - No Info, 1 - never, 2 - former, 3 - current, 4 - not current , 5 - ever
dataset['smoking_history']=pd.DataFrame(le.fit_transform(dataset['smoking_history']))
# 0- Female, 1- male, 2 - other
dataset['gender'] = pd.DataFrame(le.fit_transform(dataset['gender']))
```

```python
dataset.head(3)
```

```python
#Spliting data into input and output data
x = dataset.iloc[:,:-1]
y = dataset['diabetes']
```

```python
#Checking for max accuracy scores that the can model can get
cross_val_score(SVC(),x,y,cv = StratifiedKFold())
```

```python
#spliting data into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1,stratify=y)
```

```python
#Training the model using the Support Vector Machine algorithm
model = SVC()
model.fit(x_train, y_train)
```

```python
#checking accuracy of model by testing and training data
model.score(x_test,y_test)*100, model.score(x_train, y_train)
```

```python
#prediction of model 
outcome = model.predict([[0,54.0,	0	,0,	0,	27.32,	6.6,	80	]])
if outcome == 0:
    print("He/She has no diabetes")
else:
    print("He/She has diabaties")
```
